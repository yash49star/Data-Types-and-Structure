Q. 8 What is a hash table, and how does it relate to dictionaries in python?
-
A hash table is a data structure that stores key-value pairs, where each key is unique and its position in the table is determined by a hash function. This function takes the key as input and produces an index, or "hash," that indicates where the associated value is stored. Hash tables allow for efficient data retrieval, insertion, and deletion, with an average time complexity of O(1).

- In Python, dictionaries are implemented using hash tables. When a key-value pair is added to a dictionary, Python calculates the hash value of the key and uses it to determine the position in the underlying hash table where the value will be stored. When a value is retrieved using its key, Python again calculates the hash value of the key and uses it to quickly locate the corresponding value in the hash table. This is why dictionaries in Python provide fast lookups, insertions, and deletions.

Q. 10 Explain why strings are immutable in python.
-
In Python, strings are immutable because this design choice offers several benefits:

- Efficiency: Since strings cannot be changed after creation, Python can optimize memory allocation and caching. Multiple identical strings can share the same memory location, reducing memory usage.

- Thread safety: Immutability makes strings inherently thread-safe. Multiple threads can access and use the same string without the risk of data corruption due to concurrent modification.

- Usability as dictionary keys: Immutable objects can be used as keys in dictionaries. If strings were mutable, their hash values could change, making them unsuitable as dictionary keys.

- Preventing unintended modification: Immutability ensures that a string's value remains constant throughout its lifetime, preventing accidental or unintended modifications that could lead to bugs.

Q. 11 What advantages do dictionaries offer over lists for certain tasks?
-
Dictionaries in Python offer several advantages over lists for specific tasks, primarily related to data organization, retrieval, and efficiency. Here's a breakdown of these advantages:

1. Efficient Data Retrieval
Dictionaries provide fast data retrieval using keys. In a list, to find a specific element, it may require traversing through each item sequentially until the target is found, resulting in O(n) time complexity in the worst case. Dictionaries, on the other hand, use a hash table implementation, allowing for average-case O(1) time complexity for lookups, insertions, and deletions. This means that accessing a value in a dictionary by its key is significantly faster, especially when dealing with large datasets.

2. Key-Value Pair Organization
Dictionaries store data in key-value pairs, allowing for a more structured and intuitive way to represent relationships between data. This is particularly useful when dealing with data that has a natural association, such as storing user information (e.g., name: "John", age: 30, city: "New York") or configuration settings. Lists, in contrast, store data in a linear sequence, which may not be suitable for representing such relationships directly.

3. Flexibility and Versatility
Dictionaries can store values of any data type, and keys can be of any immutable data type (e.g., strings, numbers, tuples). This flexibility makes them suitable for a wide range of applications, including storing complex data structures, representing graphs, or implementing caches. Lists, while also versatile, do not offer the same level of structure and organization as dictionaries.

4. Built-in Methods for Manipulation
Python dictionaries provide a variety of built-in methods for easy data manipulation, such as get(), keys(), values(), and items(). These methods allow for efficient access, modification, and iteration of data within the dictionary. Lists also have built-in methods, but they are designed for different types of operations, such as sorting, appending, and inserting elements.

5. Memory Efficiency
Dictionaries are generally memory-efficient, as they only store the key-value pairs that are needed. This can be advantageous when dealing with large datasets, as it can reduce memory consumption compared to lists, which may require pre-allocation of memory.

Q. 14 How does the "in" keyword work differently for lists and dictionaries?
- The in keyword checks whether a value exists anywhere in the list.

- For Lists (and other sequences like tuples or strings):

my_list = [1, 2, 3, 4]

print(2 in my_list)      # True → 2 is in the list

print(5 in my_list)      # False → 5 is not in the list

- For Dictionaries:

The in keyword checks only for keys, not values.

my_dict = {'a': 1, 'b': 2}

print('a' in my_dict)    # True → 'a' is a key

print(1 in my_dict)      # False → 1 is a value, not a key

Q. 20 Explain the difference between a list and a dictionary in terms of data retrieval.

**- Lists:**

   1. Data is retrieved by position (index).

   2. Indexes are integers, starting from 0.

   3. Retrieval takes O(1) time if you know the index.

Example:

my_list = ['apple', 'banana', 'cherry']

print(my_list[1])  # Output: banana (retrieves item at index 1)

    To find an item by value, you must search linearly (O(n) time).

'banana' in my_list  # O(n) operation

**- Dictionaries:**

   1. Data is retrieved by key, not by index.

   2. Keys can be any immutable type (like strings, numbers, tuples).

   3. Retrieval by key is very fast — average case O(1) due to hash-based lookup.

Example:

my_dict = {'fruit1': 'apple', 'fruit2': 'banana'}

print(my_dict['fruit2'])  # Output: banana
